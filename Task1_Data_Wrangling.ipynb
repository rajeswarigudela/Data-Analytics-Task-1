```python
import pandas as pd

# Correct Superstore dataset link
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.info()
```

```python
# Create a new Date column
df['Date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'])

df.head()
```

```python
# Drop unnecessary columns
df_cleaned = df.drop(['year', 'month'], axis=1)

df_cleaned.head()
```

```python
# Save cleaned dataset as CSV
df_cleaned.to_csv("cleaned_flights_data.csv", index=False)
```
